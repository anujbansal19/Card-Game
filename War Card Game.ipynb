```python
values={'Two':2,'Three':3,'Four':4,'Five':5,'Six':6,'Seven':7,
        'Eight':8,'Nine':9,'Ten':10,'Jack':11,'Queen':12,'King':13,'Ace':14}

ranks=('Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King','Ace')

suits=('Hearts','Diamonds','Clubs','Spades')
```

```python
class Card:
    def __init__(self,suit,rank):
        self.suit=suit
        self.rank=rank
        self.value=values[rank]
        
    def __str__(self):
        return self.rank+' of '+self.suit
    
```

```python
import random
```

```python
class Deck:
    
    all_cards=[]
    def __init__(self):
        
        for rank in ranks:
            for suit in suits:
                create_card=Card(suit,rank)
                self.all_cards.append(create_card)
        
    def shuffle_deck(self):
        random.shuffle(self.all_cards)
        
    def deal_one(self):
        return self.all_cards.pop()
    
    def __str__(self):
        return f'{len(self.all_cards)} cards left!'
```

```python
class Player:
    
    def __init__(self,name):
        
        self.current_cards=[]
        self.name=name
            
    def remove_card(self):
        return self.current_cards.pop()
    
    def add_cards(self,new_cards):
        if type(new_cards) == type([]):
            self.current_cards.extend(new_cards)
        else:
            self.current_cards.append(new_cards)
    
    def __str__(self):
        return f'Player {self.name} has {len(self.current_cards)} card(s).'
```

```python
player_one = Player('One')
player_two = Player('Two')

newDeck=Deck()
newDeck.shuffle_deck()
```

```python
for i in range(26):
    player_one.add_cards(newDeck.deal_one())
    player_two.add_cards(newDeck.deal_one())
```

```python
game_on=True
round_num=0
```

```python
while game_on:
    
    round_num+=1
    print(f'Round {round_num} : ')
    
    if len(player_one.current_cards)==0:
        print('Player One has no cards!\nPlayer Two has won!')
        game_on=False
        break
        
    if len(player_two.current_cards)==0:
        print('Player Two has no cards!\nPlayer One has won!')
        game_on=False
        break
        
    player1_cards=[]
    player1_cards.append(player_one.remove_card())
    
    player2_cards=[]
    player2_cards.append(player_two.remove_card())
        
    war_on = True
        
    while war_on:
            
            if player1_cards[-1].value > player2_cards[-1].value:
                player_one.add_cards(player2_cards)
                player_one.add_cards(player1_cards)
                war_on=False
                
            elif player1_cards[-1].value < player2_cards[-1].value:
                player_two.add_cards(player2_cards)
                player_two.add_cards(player1_cards)
                war_on=False
                
            else:
                print('WAR !')
                
                if len(player_one.current_cards) < 5:
                    print('Player One has not enough cards for war!')
                    print('Player TWO wins!!')
                    game_on=False
                    break
                    
                elif len(player_two.current_cards) < 5:
                    print('Player Two has not enough cards for war!')
                    print('Player ONE wins!!')
                    game_on=False
                    break
                
                else:
                
                    for num in range(5):
                        player1_cards.append(player_one.remove_card())
                        player2_cards.append(player_two.remove_card())
```
